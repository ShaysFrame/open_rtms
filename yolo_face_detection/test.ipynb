```python
import cv2
from ultralytics import YOLO

# Load the YOLOv8 model
# model = YOLO('yolov8n.pt')  # Use your model path
# Use your model path
model = YOLO(
    '/Users/shay/Dev/Projects/project_open_rtms/open_rtms/assets/face/yolov8n-face.pt')

# Open the laptop's default camera (index 0)
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Run YOLO inference on the frame
    results = model(frame)

    # Draw detections on the frame
    annotated_frame = results[0].plot()

    # Show the frame
    cv2.imshow('YOLOv8 Detection', annotated_frame)

    # Press 'q' to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
