```python
import os
import yaml
from ultralytics import YOLO
```

```python
# Path to your data.yaml file
data_yaml_path = '/Users/shay/Dev/Projects/project_open_rtms/open_rtms/yolo_face_detection/data.yaml'

with open(data_yaml_path, 'r') as f:
    data = yaml.safe_load(f)

print(data)
```

```python
# Load a pre-trained YOLOv8n model
model = YOLO(
    '/Users/shay/Dev/Projects/project_open_rtms/open_rtms/yolo_face_detection/yolov8n.pt')
```

```python
results = model.train(data=data_yaml_path, epochs=3, imgsz=640)  # Train the model
```

```python
import matplotlib.pyplot as plt

# Run validation on the model
metrics = model.val(data=data_yaml_path)
print(metrics)

# Create visualization of validation metrics
plt.figure(figsize=(12, 8))

# Plot confusion matrix if available
if hasattr(metrics, 'confusion_matrix') and metrics.confusion_matrix is not None:
	plt.subplot(2, 2, 1)
	plt.imshow(metrics.confusion_matrix.matrix)
	plt.colorbar()
	plt.title('Confusion Matrix')
	plt.xlabel('Predicted')
	plt.ylabel('True')

# Plot precision-recall curve
plt.subplot(2, 2, 2)
if hasattr(metrics, 'pr_curve') and metrics.pr_curve is not None:
	for i, cls in enumerate(metrics.names):
		plt.plot(metrics.pr_curve[i][:, 1], metrics.pr_curve[i][:, 0], label=f'{cls} {metrics.ap[i]:.3f}')
	plt.xlabel('Recall')
	plt.ylabel('Precision')
	plt.title('Precision-Recall Curve')
	plt.legend()
else:
	plt.plot(metrics.fitness)
	plt.title('Model Fitness')

# Plot F1-Confidence curve
plt.subplot(2, 2, 3)
if hasattr(metrics, 'f1_curve') and metrics.f1_curve is not None:
	for i, cls in enumerate(metrics.names):
		plt.plot(metrics.f1_curve[i][:, 0], metrics.f1_curve[i][:, 1], label=f'{cls}')
	plt.xlabel('Confidence')
	plt.ylabel('F1 Score')
	plt.title('F1-Confidence Curve')
	plt.legend()

# Display key metrics
plt.subplot(2, 2, 4)
plt.axis('off')
metrics_text = f"""
mAP@0.5: {metrics.box.map50:.4f}
mAP@0.5:0.95: {metrics.box.map:.4f}
Precision: {metrics.box.mp:.4f}
Recall: {metrics.box.mr:.4f}
"""
plt.text(0.1, 0.5, metrics_text, fontsize=12)
plt.title('Key Metrics')

plt.tight_layout()
plt.show()
```

```python
import cv2
from ultralytics import YOLO
import time

# Load the newly trained model - adjust path to your trained model
# Look for the 'best.pt' file in the 'runs/detect/train' directory
trained_model_path = '/Users/shay/Dev/Projects/project_open_rtms/open_rtms/yolov8n-face-lindevs.pt'

try:
	# Load the trained model
	model = YOLO(trained_model_path)
	print(f"Model loaded successfully from {trained_model_path}")
	
	# Initialize webcam (MacBook camera)
	cap = cv2.VideoCapture(0)
	
	if not cap.isOpened():
		print("Could not open webcam")
		exit()
	
	# Set display window
	cv2.namedWindow("Face Detection", cv2.WINDOW_NORMAL)
	
	print("Starting detection. Press 'q' to quit.")
	
	try:
		while True:
			# Read frame from webcam
			ret, frame = cap.read()
			
			if not ret:
				print("Failed to grab frame")
				break
			
			# Perform detection
			results = model(frame)
			
			# Visualize results on the frame
			annotated_frame = results[0].plot()
			
			# Display the annotated frame
			cv2.imshow("Face Detection", annotated_frame)
			
			# Break the loop if 'q' is pressed
			if cv2.waitKey(1) & 0xFF == ord('q'):
				print("Quitting...")
				break
				
			# Add a small delay to reduce CPU usage
			time.sleep(0.01)
			
	except Exception as e:
		print(f"Error during detection: {e}")
	finally:
		# Release resources
		cap.release()
		cv2.destroyAllWindows()
		print("Camera released and windows closed")
		
except Exception as e:
	print(f"Failed to load model: {e}")
```
